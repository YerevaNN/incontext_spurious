```python
no_rotate_eval_path = 'outputs/runs/evaluate/2b1d5bf7df8f4ab18062551a/2024-11-23_03-25-29'
with_rotate_eval_path = 'outputs/runs/evaluate/2b1d5bf7df8f4ab18062551a/2024-11-23_03-47-55'
```

```python
def read_eval(eval_path):
    df = pd.read_csv(os.path.join(eval_path, 'val_train_test.csv'))
    return df.sort_values('context_class_size')
```

```python
no_rotate_eval_df = read_eval(no_rotate_eval_path)
with_rotate_eval_df = read_eval(with_rotate_eval_path)
```

```python
def parse_mean(eval_values):
    ret = []
    for x in eval_values:
        ret.append(float(x.split('±')[0]))
    return np.array(ret)


def plot_results(
    no_rotate_eval_df,
    with_rotate_eval_df,
    metric,
    save_name=None):

    fig, ax = plt.subplots(figsize=(7, 5))

    assert np.allclose(
        no_rotate_eval_df['context_class_size'],
        with_rotate_eval_df['context_class_size']
    )
    context_len = 2*np.array(no_rotate_eval_df['context_class_size'])
    
    ax.plot(context_len,
            parse_mean(no_rotate_eval_df[metric]), 
            '-o',
            label='Standard evaluation')

    ax.plot(context_len,
            parse_mean(with_rotate_eval_df[metric]),
            '-x',
            label='Evaluating with randomly rotated embeddings')
    
    # customize the plot
    if metric.find('worst_group_accuracy') != -1:
        ax.set_ylabel('Worst-group accuracy')
    elif metric.find('minority') != -1:
        ax.set_ylabel('Minority-group accuracy')
    elif metric.find('majority') != -1:
        ax.set_ylabel('Majority-group accuracy')
    else:
        ax.set_ylabel('Overall accuracy')

    ax.set_xlabel('Number of context examples')
    ax.set_xscale('log')
    ax.set_xticks(context_len)
    ax.set_xticklabels(context_len)
    ax.legend()
    fig.tight_layout()

    if save_name:
        fig.savefig(f'plots/{save_name}_{metric}.pdf')

    return fig, ax
```

```python
plot_results(
    no_rotate_eval_df=no_rotate_eval_df,
    with_rotate_eval_df=with_rotate_eval_df,
    metric='accuracy_majority',
    save_name='data_leakage'
)

plot_results(
    no_rotate_eval_df=no_rotate_eval_df,
    with_rotate_eval_df=with_rotate_eval_df,
    metric='worst_group_accuracy',
    save_name='data_leakage'
)
```
